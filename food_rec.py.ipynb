```python
#recipes['name'] = recipes['name'].str.lower()
#recipes['type'] = recipes['type'].str.lower()
#recipes['mood'] = recipes['mood'].str.lower()

#cleaning the mood column with regex
import re

def clean_mood(mood):
    return re.sub("[^a-zA-Z0-9]", " ", mood)
    
#cleaning the mood column by removing commas and converting all uppercase letters into lowercase
#def clean_mood(mood):
#    recipes['mood'] = recipes['mood'].str.replace(',', '', regex=True) #removing commas
#    recipes['mood'] = recipes['mood'].apply(str.lower) #uppercase to lowercase
#   return recipes
```

```python
recipes["clean_mood"] = recipes["mood"].apply(clean_mood)
```

```python
recipes
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2))

tfidf = vectorizer.fit_transform(recipes["clean_mood"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(mood):
    mood = clean_mood(mood)
    query_vec = vectorizer.transform([mood])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -5)[-5:]
    results = recipes.iloc[indices][::-1]
    return results
```

```python
import ipywidgets as widgets

from IPython.display import display

layout = widgets.Layout(width='auto', height='40px')

button = widgets.Button(
    description='Enter Mood: happy, sad, anxious, hungry, angry, tired, after a good cry',
    disabled=False,
    display='flex',
    flex_flow='column',
    align_items='stretch', 
    layout = layout
)

mood_name_input = widgets.Text(value = "", disable = False)

recommendation_list = widgets.Output()

def on_type(data):
    with recommendation_list:
        recommendation_list.clear_output()
        mood = data["new"]
        if len(mood) > 0:
            results = search(mood)
            #recipe_link = results.iloc[:-1] #to check all columns of the results
            recipe_link = results.loc[:, ["name", "recipe link"]]
            #display(search(mood))
            display(recipe_link)
            
mood_name_input.observe(on_type, names = "value")

display(button, mood_name_input, recommendation_list)
```
